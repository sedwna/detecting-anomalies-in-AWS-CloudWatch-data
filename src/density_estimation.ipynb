```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

sb.set_theme(style='dark')
```

```python
# load data
df = pd.read_csv("../data/elb_request_count_8c0756.csv")
df.head()
```

```python
df.dtypes
```

```python
# conver timestamp to datetime type
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.dtypes
```

```python
df.describe()
```

```python
# plot the parrt of data frame
plt.figure(figsize=(14,5))
# _ = sb.lineplot(data=df,x='timestamp',y='value')
_ = sb.lineplot(data=df[df['timestamp'] >= '2014-04-23'],x='timestamp',y='value')
```

```python
# create a timeseris 
window_size = 5
for i in range(window_size):
    df['prev_' + str(i+1)] = df['value'].shift(i+1)

# delete a row  with epmety data
df = df.dropna()
df
```

```python
# convert data to numpy array
X = df[['value'] + ['prev_'+ str(i+1) for i in range(window_size)]].values
```

```python
X
```

```python
from scipy.stats import multivariate_normal as mvn

# calculate the mean matrix in each column
loc = X.mean(axis=0)

# calculate the covariance matrix in each column with transpose
sigma = np.cov(X.T)

# get the log of pdf becuse the numbers dont be so small 
log_likelihood = mvn.logpdf(X,mean=loc,cov=sigma)
```

```python
log_likelihood
```

```python
df['ll'] = log_likelihood
```

```python
plt.figure(figsize=(14,5))
# plot the parrt of data frame
_ = sb.lineplot(data=df[df['timestamp'] >= '2014-04-23'],x='timestamp',y='value')
# plot dencitydensity estimation of data frame
_ = sb.lineplot(data=df[df['timestamp'] >= '2014-04-23'],x='timestamp',y='ll')
# plot the anomaly detected by density estimation with hyperparameter(hp)
hp = -35
_ = sb.scatterplot(data =df[(df['timestamp'] >= '2014-04-23') & (df['ll'] <= hp)],
                   x='timestamp',y='value',color='red')
```

```python
# show association betwin seris
_ = plt.matshow(sigma)
```

```python
# decomposition data
from sklearn.decomposition import PCA
pc = PCA(n_components=2).fit_transform(X)
df[['p1','p2']] = pc
```

```python
_ = sb.scatterplot(data=df,x='p1',y='p2',hue='ll',palette='coolwarm')
```
