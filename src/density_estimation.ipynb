```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt

sb.set_theme(style='white')
```

```python
df = pd.read_csv("../data/elb_request_count_8c0756.csv")
df.head()
```

```python
df.dtypes
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.dtypes
```

```python
df.describe()
```

```python
plt.figure(figsize=(14,5))
# _ = sb.lineplot(data=df,x='timestamp',y='value')
_ = sb.lineplot(data=df[df['timestamp'] >= '2014-04-23'],x='timestamp',y='value')
```

```python
window_size = 5
for i in range(window_size):
    df['prev_' + str(i+1)] = df['value'].shift(i+1)
df = df.dropna()
df
```

```python
X = df[['value'] + ['prev_'+ str(i+1) for i in range(window_size)]].values
```

```python
X
```

```python
from scipy.stats import multivariate_normal as mvn

loc = X.mean(axis=0)
sigma = np.cov(X.T)

log_likelihood = mvn.logpdf(X,mean=loc,cov=sigma)
```

```python
log_likelihood
```

```python
df['ll'] = log_likelihood
```

```python
plt.figure(figsize=(14,5))
_ = sb.lineplot(data=df[df['timestamp'] >= '2014-04-23'],x='timestamp',y='value')
_ = sb.lineplot(data=df[df['timestamp'] >= '2014-04-23'],x='timestamp',y='ll')
_ = sb.scatterplot(data =df[(df['timestamp'] >= '2014-04-23') & (df['ll'] <= -35)],
                   x='timestamp',y='value',color='red')
```

```python
_ = plt.matshow(sigma)
```

```python
from sklearn.decomposition import PCA
pc = PCA(n_components=2).fit_transform(X)
df[['p1','p2']] = pc
```

```python
_ = sb.scatterplot(data=df,x='p1',y='p2',hue='ll',palette='coolwarm')
```
